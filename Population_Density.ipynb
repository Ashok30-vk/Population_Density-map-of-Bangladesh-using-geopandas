###### Filtering row by value

```python
population_data=population_data.loc[population_data['Status']=='Dist']
```

```python
population_data
```

###### create an empty column 

```python
population_data['Name2']=''
```

```python
for index, row in population_data.iterrows():
    if '[' and ']' in row['Name']:
        start_index = row['Name'].find('[')
        end_index = row['Name'].find(']')
        population_data.loc[index, 'Name2'] = population_data.loc[index]['Name'][start_index+1: end_index]
        
    else:
        population_data.loc[index, 'Name2'] = population_data.loc[index]['Name']
```

```python
population_data
```

```python
population_data=population_data[['Name2','Population']]
```

```python
population_data.rename(columns={'Name2':'District'},inplace=True)
```

```python
population_data=population_data.reset_index(drop=True)
```
